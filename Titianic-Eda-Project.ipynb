```python
# impert all the necessary libraries 
import pandas as pd
import numpy as np
import seaborn as sns #data visualization library for generating attractive and creative graphics
import matplotlib.pyplot as plt
```

## Load the titianic Data

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
print(df.info()) #provides concise summary of our dataframe
print(df.isnull().sum()) #tells you how many missing values are in each column
```

## Statistical Summary

```python
df.describe() #gives statistical summary of the numerical columns like mean, std, min, max etc
df['Survived'].value_counts() #counts how many times each of the attribute occurs
df['Pclass'].value_counts()
```

## Key Visualizations

### 1. Survival Count

```python
sns.countplot(x='Survived', data=df) #creates a bar chart that gives a quick visual of the clas distribution
plt.title("Survival Count")
```

### Survival By Gender

```python
sns.countplot(x='Sex', hue='Survived', data=df) #Shows a stacked bar chart comparing male vs. female survival rates.
```

### Age Distribution

```python
sns.histplot(df['Age'].dropna(), kde=True) # kde=True overlays a smooth density curve for better visual analysis.
```

### Pclass vs Survival

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
```

```python
sns.heatmap(df.isnull(), cbar=False)
```

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
```
